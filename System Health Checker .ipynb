```python
import psutil

def check_disk():
    disk = psutil.disk_usage('/')
    print(f"Disk Total: {disk.total // (2**30)} GB")
    print(f"Disk Used: {disk.used // (2**30)} GB")
    print(f"Disk Free: {disk.free // (2**30)} GB")

def check_memory():
    mem = psutil.virtual_memory()
    print(f"Total RAM: {mem.total // (2**20)} MB")
    print(f"Available RAM: {mem.available // (2**20)} MB")

def check_cpu():
    print(f"CPU Cores: {psutil.cpu_count(logical=True)}")
    print(f"CPU Usage: {psutil.cpu_percent(interval=1)}%")

def main():
    print("=== System Health Checker ===")
    check_disk()
    check_memory()
    check_cpu()

if __name__ == "__main__":
    main()
```
